```python
for i in range(len(col)):
    fig = plt.figure(figsize=(10,5))
    sns.barplot(x = 'quality',y = col[i] ,data = wine)
    plt.show()
```

```python
bins = (2, 6.5, 8)
group_names = ['bad', 'good']
wine['quality'] = pd.cut(wine['quality'], bins = bins, labels = group_names)
```

```python
sns.countplot(wine['quality'])
```

```python
x = wine.drop(['quality','Id'],axis=1)
```

```python
y = wine['quality']
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=200)
```

```python
rf.fit(x_train,y_train)
pred_val = rf.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
print(classification_report(y_test,pred_val))
```

```python
print(confusion_matrix(y_test,pred_val))
```

```python
print((accuracy_score(pred_val,y_test))*100)
```
